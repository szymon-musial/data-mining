```python
from docx import Document
import PyPDF2
import pandas as pd

def find_pages_with_text(pdf_file, target_texts):
    target_pages = []
    with open(pdf_file, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page_num, page in enumerate(reader.pages, start=1):
            page_text = page.extract_text()
            lines = page_text.split("\n")
            for line_num, line in enumerate(lines, start=1):
                for target_text in target_texts:
                    if target_text in line:
                        line_to_add = ""
                        # if len(lines) > line_num:
                        #     print("new line")
                        #     line_to_add = line_to_add + '\n\n' + lines[line_num+1]
                            
                        try:
                            line_to_add = lines[line_num+1]
                            print("new line")
                        except:
                            print("An exception occurred")
                            
                            
                        target_pages.append({'Line': line, 'next_line': line_to_add, 'Page': page_num})
                        break
    return pd.DataFrame(target_pages)

# Przykładowe użycie
pdf_file = "Eksploracja Danych - dokumentacja (2).pdf"
target_texts = ["Rys ", "Rys."]

df_target_pages_rys = find_pages_with_text(pdf_file, target_texts)


# Przykładowe użycie
target_texts = ["Tabela "]

df_target_pages_tab = find_pages_with_text(pdf_file, target_texts)
```

```python
from docx import Document


def print_to_docx(df_target_pages, name):
    
    # Tworzenie nowego dokumentu
    doc = Document()

    # Dodawanie tabeli
    table = doc.add_table(rows=1, cols=3)
    table.style = 'Table Grid'  # Ustawienie stylu tabeli

    # Dodawanie nagłówków kolumn
    table.cell(0, 0).text = 'Line'
    table.cell(0, 2).text = 'Page'
    table.cell(0, 1).text = 'next_line'

    # Dodawanie danych do tabeli
    for index, row in df_target_pages.iterrows():
        line_text = row['Line']
        page_number = str(row['Page'])
        next_line = str(row['next_line'])
        row_cells = table.add_row().cells
        row_cells[0].text = line_text
        row_cells[2].text = page_number
        row_cells[1].text = next_line


    # Zapisywanie dokumentu
    doc.save(name)

print_to_docx(df_target_pages_rys, 'target_pages_rys.docx')
print_to_docx(df_target_pages_tab, 'target_pages_tab.docx')
```
